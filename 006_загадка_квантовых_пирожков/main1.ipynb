```python
from qiskit import *
from qiskit.visualization import *
from numpy import *
from numpy.linalg import *
```

```python
sim = Aer.get_backend('aer_simulator') # симулятор
```

```python
qc = QuantumCircuit(2) # 2 кубита
```

```python
# вектор удовлетворяющий всем требованиям задачи
psi = [1/2, -sqrt(3/8), -sqrt(3/8), 0]
```

```python
norm(psi)
```

```python
qc.initialize(psi)
```

```python
NR = array([[sqrt(2/5), -sqrt(3/5)], [sqrt(3/5), sqrt(2/5)]])
```

```python
# qc.unitary(NR, 0)
qc.unitary(NR, 1)
```

```python
qc.save_statevector()
qc.draw(output='mpl', initial_state=True)
```

```python
result = sim.run(qc).result()
counts = result.get_counts()
plot_histogram(counts)

# у Люси и Рекарод не вкусные пирожки
# у Люси не вкусный, у Рекардо вкусный пирожок
# у Люси вкусный, у Рекардо не вкусный пирожок
# Поднялось у Люси и Ркардо
```
