```python
sim = Aer.get_backend('aer_simulator') # симулятор
```

```python
qc = QuantumCircuit(1) # квантовая цепь с одним кубитом
```

```python
qc.x(0) # добавление гейта
```

```python
# начальный вектор состояний
qc.save_statevector()
```

```python
qc.draw(output='mpl', initial_state = True)
```

```python
result = sim.run(qc).result() # запуск симулятора
```

```python
psi = result.get_statevector() # вектор состояния
array_to_latex(psi, prefix="|psi> = \n")
```

```python
plot_bloch_multivector(psi)
```
