```python
# импорт библиотек
from qiskit import *
from qiskit.visualization import *
from numpy import *
from numpy.linalg import *
from qiskit.providers.aer import Aer
```

```python
sim = Aer.get_backend('aer_simulator') # симулятор
```

```python
qc_sum = QuantumCircuit(4,2)
```

```python
qc_sum.barrier()

qc_sum.x(0)
qc_sum.x(1)

qc_sum.cx(0,2)
qc_sum.cx(1,2)
qc_sum.ccx(0, 1,3)

qc_sum.barrier()
qc_sum.measure(2,0)
qc_sum.measure(3,1)

qc_sum.draw(output='mpl', initial_state = True)
```

```python
counts = sim.run(qc_sum).result().get_counts()
plot_histogram(counts)
```
