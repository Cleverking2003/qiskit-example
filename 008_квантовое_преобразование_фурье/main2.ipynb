```python
from qiskit import *
from qiskit.visualization import *
from numpy import *
from numpy.linalg import *
from matplotlib.pyplot import *
```

```python
n = 2 # амплитуда вероятности
```

```python
x = [0, 1] # исходный вектор состояния
```

```python
plot_bloch_multivector(x)
```

```python
# преобразование фурье
y = zeros(n, dtype = 'complex_')
for k in range(n):
    for j in range(n):
        y[k] = y[k] + x[j]*exp(-2*pi*1j*j*k/n)
    y[k] = y[k]/sqrt(n)
    
```

```python
plot_bloch_multivector(y)
```
